```python
plt.figure(figsize = (10,4))
a = sns.countplot(x = 'Age', data = df)
plt.title("Age")
```

```python
plt.figure(figsize = (15,7))
a = sns.countplot(x = 'Country', data = df)
plt.title("Country")
```

```python
plt.figure(figsize = (15,7))
a = sns.countplot(x = 'TotalLikes', data = df)
plt.title("Total Likes")
```

```python
plt.figure(figsize = (15,7))
a = sns.countplot(x = 'UsageDuraiton', data = df)
plt.title("Usage Duration")
```

```python
d = df.groupby(['Country'],as_index = False)['Age'].count().sort_values(by='Age', ascending = False)
sns.barplot(x = 'Age', y = 'Country', data = d)
```

```python
c = df.groupby(['UsageDuraiton'], as_index=False)['Age'].count().sort_values(by='Age', ascending=False)
sns.barplot(x='Age', y='UsageDuraiton', data=c)
```

```python
b =df.groupby(['Country'], as_index=False)['TotalLikes'].count().sort_values(by='TotalLikes', ascending=False)
sns.barplot(x='TotalLikes', y='Country', data=b)
```

```python
# Correct the spelling of 'UsageDuration'
v = df.groupby(['UsageDuraiton', 'Country'], as_index=False)['Age'].count().sort_values(by='Age', ascending=False)

# Plotting correctly with 'UsageDuration' and 'Country'
sns.catplot(x='Age', y='UsageDuraiton', hue='Country', data=v)
plt.show()
```

This analysis examines user engagement based on age, country, and interactions (total likes) using a dataset with 5 columns: UserId, UsageDuraiton, Age, Country, and TotalLikes.

**Likes:** Younger users (18‚Äì25) and old age (50-55) receive more likes on average.

**Country Engagement:** Users from the Turkey and Singapore have the highest UsageDuration.

# 4. **Conclusion:**
